### Cleaning the data

##### Checking the missing values

#### Feature engineering

```python
df.dtypes
```

```python
df.head()
```

#### Converting Duration time into minutes

```python
df["to_M"]=df["Duration"].str.replace("h","*60").str.replace(" ","+").str.replace("m","*1").apply(eval)
```

```python
df["to_M"]=df["to_M"].astype(int)
```

```python
df.head()
```

#### Adding new column date_of_juorney(D_o_J) with time

```python
df["D_o_J"]=df["Date_of_Journey"]+" "+df["Dep_Time"]
```

```python
df.head()
```

#### Chaging the dtype of D_o_j in datetime

```python
from datetime import datetime as dtt
```

```python
NEW=[]
for i in df["D_o_J"]:
    TP=dtt.strptime(i,"%d/%m/%Y %H:%M")
    NEW.append(TP)
```

```python
df["NEW_date"]=NEW
```

#### Now we have new datetime column with named NEW_date in dateformat

```python
df.head(3)
```

```python
df.drop(columns="D_o_J",inplace=True)
```

#### Now we add total duration of journey with New_date

```python
from datetime import timedelta as td
```

```python
n_dep_T=[]
for i in range(df.shape[0]):
    lk=int(df["to_M"][i])
    Jrd=df["NEW_date"][i]+td(minutes=lk)
    n_dep_T.append(Jrd)
```

```python
df["To_TI_OF_J"]=n_dep_T
```

```python
df.drop(columns="NEW_date",inplace=True)
```

```python
df.head(3)
```

#### Now we corret arrival time

```python
def A(a):
    return a.strftime("%H:%M %m%b")
```

```python
df["New_Arival_Time"]=list(map(A,df["To_TI_OF_J"]))
```

```python
df.head(3)
```

#### Now drop the irrelevate columns

```python
df.drop(columns=["To_TI_OF_J"],inplace=True)
```

#### Rearrange the columns

```python
df.columns
```

```python
df=df[['Airline', 'Date_of_Journey', 'Source', 'Destination', 'Route',
       'Dep_Time', 'Arrival_Time',"New_Arival_Time", 'Duration',"to_M", 'Total_Stops',
       'Additional_Info', 'Price']]
```

```python
df
```

```python
df["Airline"]=df["Airline"].str.replace("Multiple carriers Premium economy",
                          "Multiple carriers").str.replace('Jet Airways Business'
                                                           ,'Jet Airways').str.replace('Vistara Premium economy','Vistara')
```
